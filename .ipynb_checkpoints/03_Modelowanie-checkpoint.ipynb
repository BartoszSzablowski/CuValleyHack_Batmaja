## Uczenie - Modelowanie - Ewaluacja - Zapisanie

Jako model wykorzystany będzie **Random Forest**, ponieważ zależy nam na wyjaśnialności tego modelu.

**Hiperparametry** do sprawdzenia dla modelu:

```python
param_random_forest = {
    'regressor__max_depth': [10, 15, 20, 30],
    'regressor__n_estimators': [50],
    # wybór obserwacji w liściu i do splitu w oparciu o rozmiar zbioru
    'regressor__min_samples_leaf': [int(dataset.shape[0]*i) for i in [0.001, 0.005, 0.01, 0.02, 0.05, 0.1]],
    'regressor__min_samples_split': [int(dataset.shape[0]*i) for i in [0.002, 0.005, 0.01, 0.02, 0.05, 0.1]],
    'regressor__criterion': ['mse']
}
param_random_forest
```

##### Funkcja do ewaluacji modelu

```python
def model_evaluation(model, Xtrain, ytrain, Xtest, ytest):
    def fit_scatter_plot(X, y, set_name):
        y_fitted_values = model.predict(X)
        xmin = y.min()
        xmax = y.max()
        plt.scatter(x = y_fitted_values, y = y, alpha=0.25)
        x_line = np.linspace(xmin, xmax, 10)
        y_line = x_line
        plt.plot(x_line, y_line, 'r--')
        plt.axhline(0, color="black", linestyle="--")
        plt.xlabel('Prediction')
        plt.ylabel('True Value')
        plt.title(f'Plot of predicted values versus true values - {set_name} set')
    
    def plot_of_residuals(X, y, set_name):
        errors = model.predict(X) - np.reshape(np.array(y), (-1))
        plt.scatter(x = y, y = errors, alpha=0.25)
        plt.axhline(0, color="r", linestyle="--")
        plt.xlabel('True Value')
        plt.ylabel('Residual')
        plt.title(f'Plot of residuals - {set_name} set')
        
    def hist_of_residuals(X, y, set_name):
        errors = model.predict(X) - np.reshape(np.array(y), (-1))
        plt.hist(errors, bins = 100)
        plt.axvline(errors.mean(), color='k', linestyle='dashed', linewidth=1)
        plt.title(f'Histogram of residuals - {set_name} set')
    
    fig = plt.figure(figsize = (16, 10))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    
    ax = fig.add_subplot(3, 2, 1)
    fit_scatter_plot(X = Xtrain, y = ytrain, set_name = 'train')
    
    ax = fig.add_subplot(3, 2, 2)
    fit_scatter_plot(X = Xtest, y = ytest, set_name = 'test')
    
    ax = fig.add_subplot(3, 2, 3)
    plot_of_residuals(X = Xtrain, y = ytrain, set_name = 'train')
    
    ax = fig.add_subplot(3, 2, 4)
    plot_of_residuals(X = Xtest, y = ytest, set_name = 'test')    

    ax = fig.add_subplot(3, 2, 5)
    hist_of_residuals(X = Xtrain, y = ytrain, set_name = 'train')   

    ax = fig.add_subplot(3, 2, 6)
    hist_of_residuals(X = Xtest, y = ytest, set_name = 'test')
    
    plt.show()
    
    print(f'RMSE train: {sqrt(mean_squared_error(ytrain, model.predict(Xtrain)))}')
    print(f'RMSE test:  {sqrt(mean_squared_error(ytest, model.predict(Xtest)))}')
    
    # Feature Importance
    importance = model._final_estimator.feature_importances_
    df_feature_importance=importance.argsort()
    df_feature_importance=pd.DataFrame({
        'column':Xtrain.columns[df_feature_importance],
        'importance':importance[df_feature_importance]
    })
    df_feature_importance = df_feature_importance[df_feature_importance['importance']>=0.01].copy().reset_index(drop=True)
    plt.figure(figsize=(16, 4))
    plt.barh(df_feature_importance['column'][-10:], df_feature_importance['importance'][-10:])
    plt.tick_params(axis='both', labelsize=10)
    plt.title('Model Feature Importance', size=20)
    plt.xlabel(' ', size=15)
    plt.tight_layout()
    plt.show()
```

##### Funkcja do uczenia modelu:

```python
def train_model(X_tr, X_test, y_tr, y_test, name, params=param_random_forest):
    rf_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('regressor', RandomForestRegressor(n_jobs=-1))])

    rf_gridsearch = GridSearchCV(rf_pipeline, params, 
                                 cv=TimeSeriesSplit(n_splits=5).split(X_tr), n_jobs=-1,
                                 scoring = 'neg_mean_squared_error')
    rf_gridsearch.fit(X_tr, y_tr)  
    print("\nNajlepszy rezultat na zbiorach walidacyjnych: ", rf_gridsearch.best_score_,
          "\nNajlepsze hiperparametry: ", rf_gridsearch.best_params_, sep="\n")
    
    rf_model = rf_gridsearch.best_estimator_
    # Zapisanie modelu
    dump(rf_model, f'modele/model_{name}.pkl', compress=True)
    dump(rf_model.get_params(), f'modele/hyperparameters_{name}.pkl', compress=True)
    # Ocena modelu
    model_evaluation(model=rf_model, Xtrain=X_tr, ytrain=y_tr, Xtest=X_test, ytest=y_test)
    
    return rf_model
```
