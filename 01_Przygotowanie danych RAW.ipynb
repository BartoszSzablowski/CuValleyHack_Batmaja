```python
grupa_wanien_dict = {}

for grupa_wanien in range(1, 97):
    if (grupa_wanien >= 3)&(grupa_wanien <= 8):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 1,
                                           'Obwod_pradowy': 1}
    elif (grupa_wanien <= 2)|((grupa_wanien >= 9)&(grupa_wanien <= 16)):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 2,
                                           'Obwod_pradowy': 1}
    elif (grupa_wanien >= 17)&(grupa_wanien <= 24):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 3,
                                           'Obwod_pradowy': 2}
    elif (grupa_wanien >= 25)&(grupa_wanien <= 32):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 4,
                                           'Obwod_pradowy': 2}
    elif (grupa_wanien >= 33)&(grupa_wanien <= 40):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 5,
                                           'Obwod_pradowy': 3}
    elif (grupa_wanien >= 41)&(grupa_wanien <= 48):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 6,
                                           'Obwod_pradowy': 3}
    elif (grupa_wanien >= 49)&(grupa_wanien <= 56):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 7,
                                           'Obwod_pradowy': 4}
    elif (grupa_wanien >= 57)&(grupa_wanien <= 64):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 8,
                                           'Obwod_pradowy': 4}
    elif (grupa_wanien >= 65)&(grupa_wanien <= 72):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 9,
                                           'Obwod_pradowy': 5}
    elif (grupa_wanien >= 73)&(grupa_wanien <= 80):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 10,
                                           'Obwod_pradowy': 5}
    elif (grupa_wanien >= 81)&(grupa_wanien <= 88):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 11,
                                           'Obwod_pradowy': 6}
    elif (grupa_wanien >= 89)&(grupa_wanien <= 96):
        grupa_wanien_dict[grupa_wanien] = {'Grupa_wanien': grupa_wanien, 
                                           'Obieg_cyrkulacyjny_elekrolitu': 12,
                                           'Obwod_pradowy': 6}

grupa_wanien = pd.DataFrame.from_dict(grupa_wanien_dict, orient='index')
grupa_wanien
```

### Połączenie plików w oparciu o przynależność do grupy wanien, biegu cyrkulacyjnego czy obwodu prądu.

```python
def add_missing_dates(df, missing_dates):
    df_prepared = df.copy()
    df_prepared.set_index('Data', inplace=True)
    df_prepared = df_prepared.reindex(missing_dates)
    df_prepared = df_prepared.reset_index()
    df_prepared.rename({'index': 'Data'}, axis=1, inplace=True)
    return df_prepared
```

```python
def stack_dataset(df, name, typ_grupy='Obieg_cyrkulacyjny_elekrolitu'):
    df_prepared = df.copy()
    df_prepared.set_index('Data', inplace=True)
    df_prepared = df_prepared.stack(dropna=False).reset_index(level=-1).rename({'level_1': typ_grupy,
                                                                                0: name}, axis=1)
    if typ_grupy=='Obieg_cyrkulacyjny_elekrolitu':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[6:].astype(int)
    elif typ_grupy=='Obwod_pradowy':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[6:].astype(int)
    elif typ_grupy=='Grupa_wanien':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[6:].astype(int)
    elif typ_grupy=='sklad':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[5:].astype(int)
        df_prepared.rename({typ_grupy: 'Obieg_cyrkulacyjny_elekrolitu'}, axis=1, inplace=True)
    elif typ_grupy=='tiomocznik':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[20:-6].astype(int)
        df_prepared.rename({typ_grupy: 'Obieg_cyrkulacyjny_elekrolitu'}, axis=1, inplace=True)
    elif typ_grupy=='nacl':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[14:-6].astype(int)
        df_prepared.rename({typ_grupy: 'Obieg_cyrkulacyjny_elekrolitu'}, axis=1, inplace=True)
    elif typ_grupy=='klej':
        df_prepared[typ_grupy] = df_prepared[typ_grupy].str[14:-6].astype(int)
        df_prepared.rename({typ_grupy: 'Obieg_cyrkulacyjny_elekrolitu'}, axis=1, inplace=True)
        
    df_prepared = df_prepared.reset_index()
    return df_prepared
```

```python
###### Dodanie temperatury - zbiorniki naporowe oraz cyrkulacyjne
# Będą to pliki, od których wychodzimy - w pierwszym kroku dodajemy brakujące dane
# Oba zaczynają i kończą się w tym samym czasie
new_dates = pd.date_range(start=np.min([temp_naporowe['Data'].min(),
                                        temp_cyrkulacyjne['Data'].min()]), 
                          end=np.max([temp_naporowe['Data'].max(),
                                      temp_cyrkulacyjne['Data'].max()]), 
                          freq='1H')
temp_naporowe_full_days = add_missing_dates(temp_naporowe, new_dates)
temp_cyrkulacyjne_full_days = add_missing_dates(temp_cyrkulacyjne, new_dates)
                          

# Połączenie obu zbioru danych i zapisanie do ramki danej, która będzie podstawą dalszej analizy
process_dataset = pd.merge(stack_dataset(df=temp_naporowe_full_days, name='Temp_naporowe'),
                            stack_dataset(df=temp_cyrkulacyjne_full_days, name='Temp_cyrkulacyjne'),
                            how='inner', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])

###### Dodanie grupy wanien oraz obwodów prądowych
process_dataset = pd.merge(process_dataset, grupa_wanien, how='left', on='Obieg_cyrkulacyjny_elekrolitu')
process_dataset = process_dataset[['Data', 'Grupa_wanien', 'Obieg_cyrkulacyjny_elekrolitu', 'Obwod_pradowy',
                                     'Temp_naporowe', 'Temp_cyrkulacyjne']]
process_dataset.sort_values(by=['Data', 'Grupa_wanien', 'Obieg_cyrkulacyjny_elekrolitu', 'Obwod_pradowy'], inplace=True)

###### Dodanie poziomu elektrolitu- zbiorniki naporowe oraz cyrkulacyjne
process_dataset = pd.merge(process_dataset, stack_dataset(df=poziom_naporowe, name='Poziom_naporowe'),
                            how='left', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])
process_dataset = pd.merge(process_dataset, stack_dataset(df=poziom_cyrkulacyjne, name='Poziom_cyrkulacyjne'),
                            how='left', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])

###### Dodanie informacji o komponowaniu składu elektrolitu
objetosc_przeplyw_prepared = objetosc_przeplyw.copy()
objetosc_przeplyw_prepared.rename({'obieg': 'Obieg_cyrkulacyjny_elekrolitu'}, axis=1, inplace=True)
objetosc_przeplyw_prepared['Data'] = pd.to_datetime(objetosc_przeplyw_prepared['Data'])
objetosc_przeplyw_prepared['Data'] = objetosc_przeplyw_prepared['Data'] + dt.timedelta(hours = 10)
process_dataset = pd.merge(process_dataset, objetosc_przeplyw_prepared,
                            how='left', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])
# Uzupełnienie braków wartością poprzedzającą na poziomie obiegu cyrkulacyjnego - wpisywane raz na dobę
for unique_obieg in process_dataset['Obieg_cyrkulacyjny_elekrolitu'].unique():
    process_dataset.loc[process_dataset['Obieg_cyrkulacyjny_elekrolitu']==unique_obieg,
        ['kwas [m3]', 'Kondensat [m3]', 'Elektrolit [m3]', 
         'NaCl [kg]']] = process_dataset.loc[process_dataset['Obieg_cyrkulacyjny_elekrolitu']==unique_obieg, 
                            ['kwas [m3]', 'Kondensat [m3]', 'Elektrolit [m3]', 'NaCl [kg]']].fillna(method='ffill')

###### Dodanie natezenia pradu na obwodzie pradowym
prad_prepared = prad.copy()
prad_prepared.rename({'Data/godzina': 'Data'}, axis=1, inplace=True)
prad_prepared['Data'] = pd.to_datetime(prad_prepared['Data'])
process_dataset = pd.merge(process_dataset, stack_dataset(df=prad_prepared, name='Natezenie_prad',
                                                            typ_grupy='Obwod_pradowy'),
                            how='left', on=['Data', 'Obwod_pradowy'])

###### Dodanie spadku napięcia na grupach wanien
spadek_napiecie_prepared = spadek_napiecie.copy()
spadek_napiecie_prepared['Data'] = pd.to_datetime(spadek_napiecie_prepared['Data'])
process_dataset = pd.merge(process_dataset, stack_dataset(df=spadek_napiecie_prepared, name='Spadek_napiecie',
                                                            typ_grupy='Grupa_wanien'),
                            how='left', on=['Data', 'Grupa_wanien'])

###### Dodanie składu chemicznego elektrolitu
for sklad_elektrolitu_dataframe, sklad_elektrolitu_name in zip(sklad_elektrolitu_dataframes, sklad_elektrolitu_names):
    #################################### Uzupełenienie braków zerem
    # Z otrzymanej wiadomosci na Slack wynika z tego, ze w danym terminie Obieg nie byl wlaczony
    sklad_elektrolitu_dataframe.loc[:, 'Obieg1':] = sklad_elektrolitu_dataframe.loc[:, 'Obieg1':].fillna(0)
    sklad_elektrolitu_dataframe = sklad_elektrolitu_dataframe.loc[:, :'Obieg12']
    sklad_elektrolitu_dataframe.rename({'data_analizy': 'Data'}, axis=1, inplace=True)
    sklad_elektrolitu_dataframe = stack_dataset(df=sklad_elektrolitu_dataframe, name=sklad_elektrolitu_name,
                                                typ_grupy='sklad')
    # badanie zazwyczaj ma miejsce o godzinie 6:00 - dodaenie tej godziny do zmiennej 'Data'
    sklad_elektrolitu_dataframe['Data'] = sklad_elektrolitu_dataframe['Data'] + dt.timedelta(hours = 6)
    process_dataset = pd.merge(process_dataset, sklad_elektrolitu_dataframe,
                                how='left', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])
    #################################### Założenie, że w przyszłości skład elektrod się nie zmienia do następnego badania
    for unique_obieg in process_dataset['Obieg_cyrkulacyjny_elekrolitu'].unique():
        process_dataset.loc[process_dataset['Obieg_cyrkulacyjny_elekrolitu']==unique_obieg,
            sklad_elektrolitu_name] = process_dataset.loc[process_dataset['Obieg_cyrkulacyjny_elekrolitu']==unique_obieg,
                                        sklad_elektrolitu_name].fillna(method='ffill')

###### Dodanie przepływu inhibitorów
przeplyw_prepared = przeplyw.copy()
# Funkcja do dodawania godziny do daty
def add_hours(row):
    return dt.timedelta(hours = row['GODZINA'])
przeplyw_prepared['Data'] = przeplyw_prepared['Data'] + przeplyw_prepared.apply(add_hours, axis=1)
przeplyw_prepared.drop(['GODZINA'], axis=1, inplace=True)
# Tiomocznik
przeplyw_prepared_tiomocznik = pd.concat([przeplyw_prepared['Data'],
                                          przeplyw_prepared.filter(regex='^Tiomocznik', axis=1).copy()], axis=1)
przeplyw_prepared_tiomocznik = stack_dataset(df=przeplyw_prepared_tiomocznik, name='Tiomocznik', typ_grupy='tiomocznik')
# NaCl
przeplyw_prepared_NaCL = pd.concat([przeplyw_prepared['Data'],
                                    przeplyw_prepared.filter(regex='^NaCl', axis=1).copy()], axis=1)
przeplyw_prepared_NaCL = stack_dataset(df=przeplyw_prepared_NaCL, name='NaCl', typ_grupy='nacl')
# Klej
przeplyw_prepared_klej = pd.concat([przeplyw_prepared['Data'],
                                    przeplyw_prepared.filter(regex='^Klej', axis=1).copy()], axis=1)
przeplyw_prepared_klej = stack_dataset(df=przeplyw_prepared_klej, name='Klej', typ_grupy='klej')

# Połączenie ramek danych - tiomocznik, NaCL, klej
przeplyw_prepared = reduce(lambda df_left, df_right: pd.merge(df_left, df_right, 
                                                              how='outer', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu']),
                           [przeplyw_prepared_tiomocznik, przeplyw_prepared_NaCL, przeplyw_prepared_klej])
przeplyw_prepared.sort_values(by=['Data', 'Obieg_cyrkulacyjny_elekrolitu'], inplace=True)
# Nie uzupełniamy wartości null dla Tiomocznika, ponieważ docelowo nie będziemy korzystać z 1 Obiegu
przeplyw_prepared.reset_index(drop=True, inplace=True)
# Połączenie z process_dataset
process_dataset = pd.merge(process_dataset, przeplyw_prepared,
                            how='left', on=['Data', 'Obieg_cyrkulacyjny_elekrolitu'])

###### Usunięcie obiegu nr 1
process_dataset = process_dataset.loc[process_dataset['Obieg_cyrkulacyjny_elekrolitu']!=1, :]
process_dataset.reset_index(drop=True, inplace=True)

process_dataset
```

```python
quality_dataset = grupy_produkcyjne.copy()
# Wybór Trybu produkcyjnego i następnie usunięcie tej zmiennej
quality_dataset = quality_dataset.loc[quality_dataset['TRYB']=='PROD', :]
quality_dataset.drop(['TRYB'], axis=1, inplace=True)
quality_dataset.reset_index(drop=True, inplace=True)

# Dodanie Źródła wytopu oraz liczby wytopów - mamy okreslone podgrupy, unikalne dla tego zbioru - zrobimy to warunkiem
quality_dataset['ZRODLO WYTOPU'] = quality_dataset['ZRODLO WYTOPU'].str.split()
def add_source_number(row):
    if len(row['ZRODLO WYTOPU'])==4:
        return 1, int(row['ZRODLO WYTOPU'][1]), 1, int(row['ZRODLO WYTOPU'][3])
    elif len(row['ZRODLO WYTOPU'])==2:
        if row['ZRODLO WYTOPU'][0] == 'HMG':
            return 1, int(row['ZRODLO WYTOPU'][1]), 0, 0
        elif row['ZRODLO WYTOPU'][0] == 'HL':
            return 0, 0, 1, int(row['ZRODLO WYTOPU'][1])

quality_dataset['HMG'], quality_dataset['HMG_Liczba'], quality_dataset['HL'], quality_dataset['HL_Liczba'] = zip(
        *quality_dataset.apply(add_source_number, axis=1))
quality_dataset.drop(['ZRODLO WYTOPU'], axis=1, inplace=True)

###### Dodanie obiegu cyrkulacyjnego elektrolitu oraz obwodu prądowego
quality_dataset = pd.merge(quality_dataset, grupa_wanien, how='left', on='Grupa_wanien')
# Funkcja, która przesuwa pozycje cech
def move_column_inplace(df, col, pos):
    col = df.pop(col)
    df.insert(pos, col.name, col)
move_column_inplace(df=quality_dataset, col='Obieg_cyrkulacyjny_elekrolitu', pos=1)
move_column_inplace(df=quality_dataset, col='Obwod_pradowy', pos=2)

###### Dodanie zmiennych zagregowanych z analizy Anody
dane_produkcji_katod_prepared = dane_produkcji_katod.copy()
# Uzupełnienie wartości następnych poprzednią (taka jest struktura pliku) dla zmiennych grupujących
dane_produkcji_katod_prepared[['Grupa_wanien', 'DATA ZAŁĄCZENIA GRUPY', 'DATA WYŁĄCZENIA GRUPY', 'TRYB', 
    'CYKL']] = dane_produkcji_katod_prepared[['Grupa_wanien', 'DATA ZAŁĄCZENIA GRUPY', 'DATA WYŁĄCZENIA GRUPY', 'TRYB', 
                                              'CYKL']].fillna(method='ffill')

# Wybór Trybu produkcyjnego i następnie usunięcie tej zmiennej
dane_produkcji_katod_prepared = dane_produkcji_katod_prepared.loc[dane_produkcji_katod_prepared['TRYB']=='PROD', :]
dane_produkcji_katod_prepared.drop(['TRYB'], axis=1, inplace=True)

# Agregowanie wartości analizowanych dla anod
zmienne_analizowane_anoda = ['Anoda_ANALIZA Pb', 'Anoda_ANALIZA As', 'Anoda_ANALIZA Ni', 'Anoda_ANALIZA S',
                             'Anoda_ANALIZA Sb', 'Anoda_ANALIZA O2', 'Anoda_ANALIZA Bi',
                             'Anoda_ANALIZA Sn', 'Anoda_ANALIZA Zn', 'Anoda_ANALIZA Ag']
agg_dict = {}
for zmienna_anoda in zmienne_analizowane_anoda:
    agg_dict[zmienna_anoda] = [np.min, np.max]

dane_produkcji_katod_prepared = dane_produkcji_katod_prepared.groupby(['Grupa_wanien', 'DATA ZAŁĄCZENIA GRUPY',
                                                                    'DATA WYŁĄCZENIA GRUPY', 'CYKL']).aggregate(agg_dict)
# Połączenie nazw kolumn - obecnie MultiIndex
dane_produkcji_katod_prepared.columns = ['_'.join(col) for col in dane_produkcji_katod_prepared.columns.values]

# Dodanie zmiennej różnicy
for zmienna_anoda in zmienne_analizowane_anoda:
    dane_produkcji_katod_prepared[f'{zmienna_anoda}_diff'] = dane_produkcji_katod_prepared[f'{zmienna_anoda}_amax'] -\
                                                                dane_produkcji_katod_prepared[f'{zmienna_anoda}_amin']
    
dane_produkcji_katod_prepared = dane_produkcji_katod_prepared.reset_index()
quality_dataset_columns = quality_dataset.columns
quality_dataset = pd.merge(quality_dataset, dane_produkcji_katod_prepared,
                           on=['Grupa_wanien', 'DATA ZAŁĄCZENIA GRUPY', 'DATA WYŁĄCZENIA GRUPY', 'CYKL'], how='left')
new_columns = []
for quality_dataset_column in quality_dataset_columns:
    if quality_dataset_column in zmienne_analizowane_anoda:
        new_columns.append(quality_dataset_column)
        new_columns.append(f'{quality_dataset_column}_amin')
        new_columns.append(f'{quality_dataset_column}_amax')
        new_columns.append(f'{quality_dataset_column}_diff')
    else:
        new_columns.append(quality_dataset_column)

quality_dataset = quality_dataset[new_columns]

# Dodanie unikalnego indeksu, ktory bedzie polaczaniem z danymi procesowymi
quality_dataset.reset_index(inplace=True)
quality_dataset.rename({'index': 'index_process'}, axis=1, inplace=True)

# Czas trwania procesu
quality_dataset['Czas'] = quality_dataset['DATA WYŁĄCZENIA GRUPY'] - quality_dataset['DATA ZAŁĄCZENIA GRUPY']
quality_dataset['Czas_h'] = np.round((quality_dataset['Czas'].astype('timedelta64[s]') / (60 * 60)))

quality_dataset
```

```python
quality_dataset_index = quality_dataset.copy()
quality_dataset_index = quality_dataset_index[['DATA ZAŁĄCZENIA GRUPY', 'DATA WYŁĄCZENIA GRUPY', 'index_process', 'CYKL',
                                               'Grupa_wanien', 'Obieg_cyrkulacyjny_elekrolitu', 'Obwod_pradowy']]

# Listy, ktore beda podstawa dolaczanego zbioru danych
data_list = []
start_list = []
end_list = []
index_list = []
cykl_list = []
grupa_wanien_list = []
obieg_list = []
obwod_list = []

# Dodanie dat z zakresu pomiędzy załączeniem i wyłączeniem grupy
for index, row in quality_dataset_index.iterrows():
    # Dodajemy 30 min, ponieważ zaokraglenia sa wykonywane do najblizszej godziny zgodnie z zasadami
    # w naszym przypadku musimy zaokraglac zawsze w gore
    hours = pd.date_range(start=(row['DATA ZAŁĄCZENIA GRUPY'] + dt.timedelta(minutes = 30)).round('1H'),
                          end=(row['DATA WYŁĄCZENIA GRUPY'] + dt.timedelta(minutes = 30)).round('1H'),
                          freq='1H')
    
    for hour in hours:
        data_list.append(hour)
        start_list.append(row['DATA ZAŁĄCZENIA GRUPY'])
        end_list.append(row['DATA WYŁĄCZENIA GRUPY'])
        index_list.append(row['index_process'])
        cykl_list.append(row['CYKL'])
        grupa_wanien_list.append(row['Grupa_wanien'])
        obieg_list.append(row['Obieg_cyrkulacyjny_elekrolitu'])
        obwod_list.append(row['Obwod_pradowy'])

dict_quality_dataset_index = {'Data': data_list, 'DATA ZAŁĄCZENIA GRUPY': start_list, 'DATA WYŁĄCZENIA GRUPY': end_list,
                              'index_process': index_list, 'CYKL': cykl_list, 'Grupa_wanien': grupa_wanien_list,
                              'Obieg_cyrkulacyjny_elekrolitu': obieg_list, 'Obwod_pradowy': obwod_list}

quality_dataset_index = pd.DataFrame(dict_quality_dataset_index)
quality_dataset_index
```
