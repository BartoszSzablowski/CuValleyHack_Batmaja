# Wczytanie zbiorów danych

```python
quality_dataset = pd.read_csv('output/prepared_quality_dataset.csv', index_col=False)
quality_dataset = pd.concat([quality_dataset, pd.get_dummies(quality_dataset['CYKL'],
                                                             drop_first=False)],
                            axis=1)
quality_dataset.rename({1: 'Cykl_1', 2: 'Cykl_2'}, axis=1, inplace=True)
quality_dataset.drop(['Cykl_2'], axis=1, inplace=True)
quality_dataset
```
